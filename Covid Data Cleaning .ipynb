# COVID-19 DATA CLEANING 

#### importing the libraries

```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
```

#### importing the data

```python
df=pd.read_csv('owid-covid-data.csv')
df.head()
```

#### dropping unnecessary columns

```python
df.drop(columns=['total_vaccinations','total_deaths_per_million','new_people_vaccinated_smoothed','total_tests','total_tests_per_thousand','total_cases_per_million','total_deaths','total_cases','total_vaccinations_per_hundred',
    'people_vaccinated_per_hundred','people_fully_vaccinated_per_hundred',"new_vaccinations_smoothed_per_million",
    "tests_per_case",'new_tests_smoothed_per_thousand','tests_units','handwashing_facilities','iso_code',
    'total_boosters','total_boosters_per_hundred','female_smokers','male_smokers','cardiovasc_death_rate',
    'diabetes_prevalence','excess_mortality_cumulative_absolute','excess_mortality_cumulative_per_million',
    'new_people_vaccinated_smoothed_per_hundred','excess_mortality_cumulative','extreme_poverty','extreme_poverty','population_density'],inplace=True)
```

```python
df.info()
```

#### checking the count of null values in each column

```python
df.isnull().sum()
```

#### removing null values from the column continent

```python
df.dropna(subset=['continent'],inplace=True)
```

#### removing columns with more than 50% missing values

```python
x=len(df)*0.5
cleaned_df=df.dropna(axis=1,thresh=x)
```

#### filling in the missing values for the age groups with the median value of their respective column

```python
df['aged_65_older'].fillna(df['aged_65_older'].median, inplace=True)
df['aged_70_older'].fillna(df['aged_70_older'].median, inplace=True)
```

#### filling the null values in median age with previous values 

```python
df['median_age'].fillna(method='ffill',inplace=True)
```

#### filling missing values in the column stringency index with 0

```python
df['stringency_index'].fillna(0,inplace=True)
```

```python
df.isna().sum()
```

#### Changing the format of date

```python
df['Date of case'] = pd.to_datetime(df['date'])
```

```python
df.to_csv('cleaned_covid_data.csv',index=False)
```
